### 1. Examining and loading data
Examine "\<student\_id\>_dataset2.csv" the structure and schema. 

```python
# ===============================
# 3.1 Load Cleaned Dataset1 and Dataset2
# ===============================


# Load the datasets
dataset1 = pd.read_csv("S3859590_dataset1_solution.csv")
dataset2 = pd.read_csv("S3859590_dataset2.csv")

print("Datasets loaded successfully.")
print("\nDataset1 shape:", dataset1.shape)
print("Dataset2 shape:", dataset2.shape)


```

The column names in dataset1 were renamed to match those in dataset2. This ensured that the structure of both datasets aligned correctly before merging.

```python
# ===============================
# 3.2 Rename Columns in Dataset1 to Match Dataset2
# ===============================

# Correct mapping for renaming columns in dataset1
rename_mapping = {
    'ClothID': 'Clothing ID',         # ClothID -> Clothing ID
    'Review Title': 'Title',          # Review Title -> Title
    'Customer Rating': 'Rating',      # Customer Rating -> Rating
    'Positive Review Count': 'Positive Feedback Count',  # Positive Review Count -> Positive Feedback Count
    'Section': 'Division Name',      # Section -> Division Name
    'Department': 'Department Name', # Department -> Department Name
    'Category': 'Class Name',        # Category -> Class Name
    'Online Time': 'Active Time',    # Online Time -> Active Time
    'Cost': 'Price',                 # Cost -> Price
    'Recommended_IND': 'Recommended IND'
}

# Rename columns in dataset1 based on the mapping
dataset1.rename(columns={k: v for k, v in rename_mapping.items() if k in dataset1.columns}, inplace=True)

print("\nDataset1 Columns after renaming:")
print(dataset1.columns.tolist())
```

```python
# ===============================
# 3.3 Add Missing Columns Safely
# ===============================

# Get expected columns from dataset2
expected_columns = dataset2.columns.tolist()

# Add missing columns to dataset1 with NaN values
for col in expected_columns:
    if col not in dataset1.columns:
        dataset1[col] = pd.NA  # Add missing columns with NaN (empty values)

# Reorder columns to match dataset2's column order
dataset1 = dataset1[expected_columns]

print("\nDataset1 columns aligned to match dataset2.")
```

```python
# ===============================
# 3.4 Merge Datasets
# ===============================

# Merge datasets row-wise (vertical stacking)
integrated_dataset = pd.concat([dataset1, dataset2], ignore_index=True)

print("\nDatasets merged successfully.")
print("Merged dataset shape:", integrated_dataset.shape)
```

```python
# ===============================
# 3.5 Detect and Remove Duplicates
# ===============================

# Check for duplicates
duplicates_count = integrated_dataset.duplicated().sum()
print(f"\nNumber of duplicate rows before cleaning: {duplicates_count}")

# Remove duplicates if any
if duplicates_count > 0:
    integrated_dataset = integrated_dataset.drop_duplicates().reset_index(drop=True)
    print("Duplicates removed.")
else:
    print("No duplicates found.")

print("Final dataset shape after duplicate removal:", integrated_dataset.shape)
```

```python
# ===============================
# 3.6 Check Global Key
# ===============================

# Check if 'Clothing ID' alone is unique
if 'Clothing ID' in integrated_dataset.columns:
    is_unique = integrated_dataset['Clothing ID'].is_unique
    print("\nIs 'Clothing ID' alone a unique identifier?:", is_unique)

    # If 'Clothing ID' is not unique, create a composite global key
    if not is_unique and {'Age', 'Title'}.issubset(integrated_dataset.columns):
        integrated_dataset['Global_Key'] = integrated_dataset['Clothing ID'].astype(str) + "_" + \
                                           integrated_dataset['Age'].astype(str) + "_" + \
                                           integrated_dataset['Title'].astype(str)
        is_composite_unique = integrated_dataset['Global_Key'].is_unique
        print(f"Is composite key (Clothing ID + Age + Title) unique?: {is_composite_unique}")
        integrated_dataset = integrated_dataset.drop(columns=['Global_Key'])
else:
    print("'Clothing ID' column missing, cannot check uniqueness.")
```

```python
# ===============================
# 3.7 Save Final Integrated Dataset
# ===============================

# Save the final integrated dataset
integrated_dataset.to_csv("S3859590_dataset_integrated.csv", index=False)

print("\nFinal merged dataset saved as 'S3859590_dataset_integrated.csv'.")
```
